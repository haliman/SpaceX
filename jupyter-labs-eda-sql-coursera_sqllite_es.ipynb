## Tareas

### Tarea 1




##### Muestra los nombres de los sitios de lanzamiento únicos en la misión espacial.


```python
%%sql
select distinct Launch_Site from SPACEXTBL
```

### Tarea 2


##### Muestra 5 registros donde los sitios de lanzamiento comienzan con la cadena 'CCA'


```python
%%sql
select * from SPACEXTBL where Launch_Site like 'CCA%' limit 5 
```

### Tarea 3




##### Muestra la masa total de carga útil transportada por los propulsores lanzados por la NASA (CRS)


```python
%%sql
select Payload, sum(PAYLOAD_MASS__KG_) from SPACEXTBL where Customer like '%(CRS)%'
```

### Tarea 4




##### Muestra la masa de carga útil promedio transportada por la versión de refuerzo F9 v1.1


```python
%%sql
select Payload, avg(PAYLOAD_MASS__KG_) from SPACEXTBL where Booster_Version like '%F9 v1.1%'
```

### Tarea 5

##### Indique la fecha en la que se logró el primer aterrizaje exitoso en la plataforma de tierra.


_Sugerencia: use la función mínima_


```python
%%sql
select min(SPACEXTBL.Date) from SPACEXTBL where Landing_Outcome='Success (ground pad)'
```

### Tarea 6

##### Enumere los nombres de los propulsores que tienen éxito en naves no tripuladas y tienen una masa de carga superior a 4000 pero inferior a 6000.


```python
%%sql
select Booster_Version from SPACEXTBL where Landing_Outcome='Success (drone ship)' and PAYLOAD_MASS__KG_ between 4000 and 6000
```

### Tarea 7




##### Enumere el número total de resultados de misiones exitosas y fallidas.


```python
%%sql
select Mission_Outcome, count(*) from SPACEXTBL group by 1
```

### Tarea 8



##### Enumere todas las booster_versions que han llevado la masa máxima de carga útil, utilizando una subconsulta con una función agregada adecuada.


```python
%%sql
select Booster_Version from SPACEXTBL where PAYLOAD_MASS__KG_ = (select max(PAYLOAD_MASS__KG_) from SPACEXTBL)
```

### Tarea 9


##### Enumere los registros que mostrarán los nombres de los meses, resultados de aterrizaje fallidos en el barco no tripulado, versiones de refuerzo, sitio de lanzamiento para los meses del año 2015.

**Nota: SQLLite no admite nombres de meses. Por lo tanto, debe usar substr(Date, 6,2) como mes para obtener los meses y substr(Date,0,5)='2015' para el año.**


```python
%%sql
select substr(Date, 6, 2) as "month", Booster_Version, Launch_Site from SPACEXTBL where Landing_Outcome = 'Failure (drone ship)' and substr(Date,0,5)='2015'
```

### Tarea 10




##### Clasifique el recuento de resultados de aterrizaje (como fracaso (barco drone) o éxito (plataforma de tierra)) entre la fecha 2010-06-04 y 2017-03-20, en orden descendente.


```python
%%sql

SELECT 
    Landing_Outcome, 
    COUNT(*) AS Total
FROM SPACEXTBL
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
  AND Landing_Outcome IN ('Failure (drone ship)', 'Success (ground pad)')
GROUP BY Landing_Outcome
ORDER BY Total DESC;
```
